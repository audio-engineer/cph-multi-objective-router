```python
import networkx as nx
import osmnx as ox
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"

ox.__version__
```

```python
GRAPH_CYCLING = ox.graph.graph_from_place(
    "Copenhagen Municipality, Capital Region of Denmark, Denmark", network_type="bike"
)

GRAPH_WALKING = ox.graph.graph_from_place(
    "Copenhagen Municipality, Capital Region of Denmark, Denmark", network_type="walk"
)
```

```python
GRAPH_CYCLING = ox.project_graph(GRAPH_CYCLING)
GRAPH_WALKING = ox.project_graph(GRAPH_WALKING)

display("Cycling")
display(f"Nodes: {GRAPH_CYCLING.number_of_nodes()}")
display(f"Edges: {GRAPH_CYCLING.number_of_edges()}")
display("Walking")
display(f"Nodes: {GRAPH_WALKING.number_of_nodes()}")
display(f"Edges: {GRAPH_WALKING.number_of_edges()}")
```

```python
P = ox.graph.graph_from_place(
    "Copenhagen Municipality, Capital Region of Denmark, Denmark", network_type="bike"
)
```

```python
point_a = ox.geocode("Oldermandsvej 49, Copenhagen, Denmark")
point_b = ox.geocode("Nyhavn, Copenhagen, Denmark")

node_a = ox.distance.nearest_nodes(P, X=point_a[1], Y=point_a[0])
node_b = ox.distance.nearest_nodes(P, X=point_b[1], Y=point_b[0])

path = nx.shortest_path(P, source=node_a, target=node_b, weight="length")

ox.plot_graph_route(P, path)
```

```python
coordinates = []

for node_id in path:
    lon = GRAPH_CYCLING.nodes[node_id]["x"]
    lat = GRAPH_CYCLING.nodes[node_id]["y"]

    coordinates.append((lon, lat))

coordinates
```
